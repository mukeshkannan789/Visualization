```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style="ticks", color_codes=True)
titanic = pd.read_excel("/home/perlzuser/Mukesh_Bkup/indixuser/Documents/Mukesh/Machine_learning/titanic3.xls")
titanic.head()
```

```python
g = sns.FacetGrid(titanic, col="sex", row="survived")
g = (g.map(plt.hist, "pclass").add_legend())
```

```python
g = sns.FacetGrid(titanic, col="sex", hue="survived")
g = (g.map(plt.scatter, "age","pclass").add_legend())
```

```python
titanic['gender'] = pd.factorize(titanic.sex)[0]
g = sns.FacetGrid(titanic, col="gender")
g = g.map(plt.plot, "survived","pclass")
```
